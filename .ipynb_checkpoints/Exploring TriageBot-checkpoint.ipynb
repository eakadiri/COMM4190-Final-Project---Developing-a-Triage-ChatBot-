# Exploring TriageBot

After asssessing existing LLMs, I decided that a censored LLM, like GPT-3.5/4 would be best to utilize. This can help ensure that the bot's responses are more cautious, considerate of potential legal or ethical implications, and adhere to medical guidelines and best practices. I will be using the same prompt system.

However, I decided to first test whether medlamma2 or meditron would be better suited from the ollama server. I will be slightly changing the prompting method and combine the first and second prompt as to better replicate how a bot would be asked to help in a real-life scenario. 

# medllama2

medllama2 is a Llama 2 model to answer medical questions based on an open source medical dataset.

**Prompt:** "You are a part of a triage team in an ER. These are the areas in the ER: Green(minor ailments like a cold), Yellow(moderate emergencies like infections like, and Red(for life-threatening emergencies), and Trauma(for all trauma patients). A patient has just arrived to the ER with multiple fractured ribs and shortness of breath. Can you help the team figure out where the patient should go?"

**Response:** "Given that this patient has multiple fractured ribs, which is a 
life-threatening condition, they should be immediately transported to the 
Trauma Bay for further evaluation and treatment. The patient's shortness 
of breath could also indicate other potential issues such as pneumothorax 
or cardiac dysfunction, so this should also be investigated."

This response was good. It provided limited justification for where the patient should be sent to and other possible reasons behind. There was still a slight issue of overexplanation, but it is still under 3 sentences which is pretty short. 

# meditron 

meditron is an Open-source medical large language model adapted from Llama 2 to the medical domain.

**Prompt:** "You are a part of a triage team in an ER. These are the areas in the ER: Green(minor ailments like a cold), Yellow(moderate emergencies like infections like, and Red(for life-threatening emergencies), and Trauma(for all trauma patients). A patient has just arrived to the ER with multiple fractured ribs and shortness of breath. Can you help the team figure out where the patient should go?"

**Response:** "You are a nurse who is giving a briefing about COVID-19 to your 
colleagues. You want to make sure everyone knows what they need to know in
order to be able to care for their patients and themselves."

I asked the same prompt 3 times, and each time the response was the same. It did not take into account the question, but simply created a scenario to combat mine. Considering this, medllama2 was the clear winner. 


# TriageBot

```python
from langchain_openai import ChatOpenAI, OpenAI
from langchain_community.llms import Ollama
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.messages import BaseMessage, AIMessage
from langchain_core.pydantic_v1 import BaseModel, Field
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain.memory import ChatMessageHistory
from langchain_core.messages import AIMessage, HumanMessage
from langchain.agents import AgentExecutor
from langchain.memory import ChatMessageHistory
from langchain_community.tools.pubmed.tool import PubmedQueryRun


from operator import itemgetter
from typing import List
from IPython.display import display, Markdown
import os
from dotenv import load_dotenv

_ = load_dotenv()

#Initialize the Ollama model
llm = Ollama(
    base_url="http://10.30.16.100:11434",
    model="medllama2"
)


# Define the PubMed tool
pubmed_tool = PubmedQueryRun()


class InMemoryHistory(BaseChatMessageHistory, BaseModel):
    """In memory implementation of chat message history."""

    messages: List[BaseMessage] = Field(default_factory=list)

    def add_messages(self, messages: List[BaseMessage]) -> None:
        """Add a list of messages to the store"""
        self.messages.extend(messages)

    def clear(self) -> None:
        self.messages = []
        

# Here we use a global variable to store the chat message history.
# This will make it easier to inspect it to see the underlying results.
store = {}

def get_by_session_id(session_id: str) -> BaseChatMessageHistory:
    if session_id not in store:
        store[session_id] = InMemoryHistory()
    return store[session_id]


history = get_by_session_id("1")
history.add_message(AIMessage(content="hello"))
print(store) 

    
prompt1 = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            '''You are a part of a triage team in an ER. These are the areas in the ER: Green(minor ailments like a cold), 
            Yellow(moderate emergencies like infections like, and Red(for life-threatening emergencies), and Trauma(for all trauma patients). 
            Can you help the team figure out where the patient should go?
            ''',
        ),
        
        MessagesPlaceholder(variable_name="history"),
        ("human", "{input}"),
    ]
)

chain = prompt1 | llm 

chain_with_history = RunnableWithMessageHistory(
    chain,
    # Uses the get_by_session_id function defined in the example
    # above.
    get_by_session_id,
    input_messages_key="question",
    history_messages_key="history",
)

# Extract the medical query from the context
medical_query = "You are a part of a triage team in an ER. These are the areas in the ER: Green(minor ailments like a cold), Yellow(moderate emergencies like infections like, and Red(for life-threatening emergencies), and Trauma(for all trauma patients).  Can you help the team figure out where the patient should go? A patient has just arrived to the ER with multiple fractured ribs and shortness of breath."

# Call pubmed_tool.invoke with the extracted query
pubmed_results = pubmed_tool.invoke(medical_query)

print(chain_with_history.invoke(  # noqa: T201
    {"input": medical_query},
    config={"configurable": {"session_id": "foo"}}
))

print(pubmed_results)

while True:
  # Prompt user for new query
  user_query = input("Enter your medical question (or 'quit' to exit): ")

  # Exit loop if user enters 'quit'
  if user_query.lower() == 'quit':
    break

  # Update chat history with user query
  history.add_messages([HumanMessage(content=user_query)])

  # Call pubmed_tool (optional, based on your needs)
  pubmed_results = pubmed_tool.invoke(user_query)

  # Invoke chain with history for MedLLama2 response
  response = chain_with_history.invoke(
      {"input": user_query},
      config={"configurable": {"session_id": "foo"}}
  )
  
  # Process and print MedLLama2 response (and potentially Pubmed results)
  print(f"MedLLama2 Response: {response}")
  # ... (your code to handle and present Pubmed results)

  print(pubmed_results)

print("Exiting...")
```

I tried to use PubMed as a secondary source, however it seems it is better suited to look up symptoms, let the bot figure out where the patient should go, and allow Pubmed to speculate what the patient could be going through. 

It is also interesting that the bot is also equipping itself with the knowledge to know when something is treatable in the ER vs. when it is not, as seen with the second query:
